```python
from helpers import renderMappedTrainingData
import matplotlib.pyplot as plt
import os, csv, json
from IPython.display import display
```

```python
def display_sample(participant_number, left_or_right,sample_number):
    
    sampleDir = os.path.normpath("data2/"+ participant_number + "/feet/"+ left_or_right +"/samples/" + sample_number)
    
    # load first row of csv
    csv_file = os.path.join(sampleDir, "augmentedPressure.csv")
    with open(csv_file, mode='r') as file:
        csv_reader = csv.reader(file)    
        pressure_data = next(csv_reader)
        pressure_data = [float(cell) for cell in pressure_data]
    
    # Load points
    points_file = os.path.join(sampleDir, "augmentedPoints.json")
    with open(points_file, 'r') as file:
        data = file.read()
        points_data = json.loads(data)
    
    
    img = renderMappedTrainingData(points_data, pressure_data, 20)
    
    display(img)
    
    
    
```

```python
#printing out all participants numbers and sample numbers to use for single display mode

for i in os.listdir("data2"):
    print("Participant number: " + i) 
    for left_or_right in ["left", "right"]:
        print("Foot: " + left_or_right)
        print("Sample list: " + str(os.listdir("data2/"+ i + "/feet/"+ left_or_right +"/samples/")))
    print("-----------------")
    
```

```python
# run a single participant display

display_sample("43", "right", "546")
```

```python
# display all data for all participants and all samples will cause a crash
# display all samples for each participant instead

# test set (might change) ['46', '35', '19', '30', '31', '52', '16']

#set desired participant, write as string
participant_number = "35" 

for left_or_right in ["left", "right"]:
    print(left_or_right)
    for sample_number in os.listdir("data2/"+ participant_number + "/feet/"+ left_or_right +"/samples/"):
        print(sample_number)
        display_sample(participant_number, left_or_right,sample_number)
    

# print(os.listdir("data/19/feet/right/samples"))
```
