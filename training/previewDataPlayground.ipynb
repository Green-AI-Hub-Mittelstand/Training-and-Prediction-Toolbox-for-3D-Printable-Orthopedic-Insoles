```python
%load_ext autoreload
%autoreload 2

from helpers import renderMappedTrainingData
import matplotlib.pyplot as plt
import os, csv, json
from IPython.display import display
```

```python
def display_sample(participant_number, left_or_right,sample_number):
    
    sampleDir = os.path.normpath("data2/"+ str(participant_number) + "/feet/"+ left_or_right +"/samples/" + str(sample_number))
    
    # load first row of csv
    csv_file = os.path.join(sampleDir, "pressure.csv")
    with open(csv_file, mode='r') as file:
        csv_reader = csv.reader(file)
        pressure_data = next(csv_reader)
        pressure_data = [float(cell) for cell in pressure_data]
    
    # Load points
    points_file = os.path.join(sampleDir, "points.json")
    with open(points_file, 'r') as file:
        data = file.read()
        points_data = json.loads(data)


    # create some dummy predictions
    # points_data_predictions = []
    # for p in points_data:
    #     p2 = json.loads(json.dumps(p))
    #     coords = p2['points']
    #     p2['points'] = [coords[0]-1, coords[1]+1]
    #     points_data_predictions.append(p2)
    
    # Load points
    points_predictions_file = os.path.join(sampleDir, "predictedPoints.json")
    with open(points_predictions_file, 'r') as file:
        data = file.read()
        points_data_predictions = json.loads(data)
    
    
    img = renderMappedTrainingData(points_data, pressure_data, 25, points_data_predictions, True)
    
    display(img)
    
    
    
```
