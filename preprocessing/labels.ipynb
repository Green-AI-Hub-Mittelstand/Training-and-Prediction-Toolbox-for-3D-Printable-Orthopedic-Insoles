```python
import fitz  # PyMuPDF
import qrcode
import tempfile
import shutil, os
```

```python
class PDF(object):
  def __init__(self, pdf, size=(200,200)):
    self.pdf = pdf
    self.size = size

  def _repr_html_(self):
    return '<iframe src={0} width={1[0]} height={1[1]}></iframe>'.format(self.pdf, self.size)

  def _repr_latex_(self):
    return r'\includegraphics[width=1.0\textwidth]{{{0}}}'.format(self.pdf)
```

```python
def create_label_pdf(output_file,start=1):
    # Constants for label layout and A4 dimensions
    fmt = fitz.paper_rect("a4")
    # scale
    scale = fmt.width / 210.0

    
    num_columns = 4
    num_rows = 16
    
    top_margin_mm = 13 * scale
    bottom_margin_mm = 13 * scale
    left_margin_mm = 8 *scale
    right_margin_mm = 8 * scale
    label_width_mm = (210 * scale  - left_margin_mm - right_margin_mm) / num_columns
    label_height_mm = (297 * scale - top_margin_mm - bottom_margin_mm) / num_rows

    
    
    # Create a new PDF document
    pdf_document = fitz.open()

    
    
    # Add a new page to the document
    print(fmt)
    page = pdf_document.new_page(width=fmt.width, height=fmt.height)

    counter = -1 + start

    # Loop through rows and columns to create labels
    for row in range(num_rows):
        for col in range(0,num_columns,2):
            counter+=1



            num = counter

            x_pos_offset = 20 * scale
            y_pos_offset = 10 * scale + 3

            qr_x_offset = 3 * scale
            qr_y_offset = 2 * scale - 2
            qr_size = 14 * scale

            
            # Calculate label position
            x_pos = left_margin_mm + col * label_width_mm
            y_pos =  top_margin_mm + row * (label_height_mm)

            


            # Draw a rectangle for the label
            #page.draw_rect(fitz.Rect(x_pos, y_pos, x_pos + label_width_mm, y_pos + label_height_mm))

            # generate the qr code file
            # Generate QR code
            qr = qrcode.QRCode(version=1, error_correction=qrcode.constants.ERROR_CORRECT_L, box_size=10, border=4)
            qr.add_data(str(num))  # Your URL here
            qr.make(fit=True)
            
            # Create an image from the QR code
            qr_img = qr.make_image(fill_color="black", back_color="white")
            qr_img.save("qr_code.png")  # Save the QR code image


            # Add text or content to the label (you can customize this part)
            page.insert_text((x_pos + x_pos_offset, y_pos + y_pos_offset), str(num), fontsize=8 * scale)

            
            
            
            image_rect = fitz.Rect(x_pos + qr_x_offset, y_pos + qr_y_offset, x_pos + qr_x_offset + qr_size, y_pos + qr_y_offset + qr_size)  # Adjust the coordinates and size as needed
            page.insert_image(image_rect, filename="qr_code.png")
            
            # Calculate label position
            x_pos = left_margin_mm + (col+1) * label_width_mm
            y_pos = top_margin_mm + row * (label_height_mm)

            # Draw a rectangle for the label
            #page.draw_rect(fitz.Rect(x_pos, y_pos, x_pos + label_width_mm, y_pos + label_height_mm))

            # Add text or content to the label (you can customize this part)
            
            page.insert_text((x_pos + x_pos_offset, y_pos + y_pos_offset), str(num), fontsize=8 * scale)

            image_rect = fitz.Rect(x_pos + qr_x_offset, y_pos + qr_y_offset, x_pos + qr_x_offset + qr_size, y_pos + qr_y_offset + qr_size)  # Adjust the coordinates and size as needed
            page.insert_image(image_rect, filename="qr_code.png")
            

            os.remove("qr_code.png")

    # Save the PDF to the specified output file
    pdf_document.save(output_file)
    pdf_document.close()
```

```python
start = 56
num_sheets = 6

for i in range(0,num_sheets):
    start_new = i*32 + start
    print(start_new)
    create_label_pdf("labels-%s.pdf" % start_new,start_new)    




#PDF("labels-3.pdf", (1000,1000))
```
